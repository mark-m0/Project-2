```python
columns_to_drop = ['Unnamed: 0', 'wfo',
       'injuries_direct', 'injuries_indirect',
       'deaths_direct', 'deaths_indirect', 'damage_property', 'damage_crops',
       'magnitude', 'magnitude_type', 'flood_cause', 'tor_f_scale',
       'tor_length', 'tor_width', 'tor_other_wfo', 'location_index',
       'event_range', 'event_azimuth', 'reference_location',
       'event_point']
columns_to_drop
```
